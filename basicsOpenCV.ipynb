```
import cv2
import imutils
from google.colab.patches import cv2_imshow
```

```
image = cv2.imread('/content/drive/My Drive/IMG-20190303-WA0152.jpg')
(h,w,d) = image.shape
print("height={}, width={}, depth={}".format(h,w,d))


cv2_imshow(image)
cv2.waitKey(0)
```

```
(b,g,r) = image[152,19]
print("r={},g={},b={}".format(r,g,b))
```

```
roi = image[260:490,319:480]
cv2_imshow(roi)
cv2.waitKey(0)
```

```
resized_image = cv2.resize(image,(229,229))
cv2_imshow(resized_image)
cv2.waitKey(0)
```

```
new_height = int(image.shape[0]/4)
print(image.shape)
print(new_height)
dim = (int((w*new_height)/h),new_height)
new_resize = cv2.resize(image,dim)
print(new_resize.shape)
cv2_imshow(new_resize)
cv2.waitKey(0)
```

```
#using imutils to resize the image
new_resized1 = imutils.resize(image,height=229)
cv2_imshow(new_resized1)
cv2.waitKey(0)
```

```
new_resized1.shape
```

```
image1 = new_resized1
(h1,w1) = image1.shape[:2]
center = (w1//2,h1//2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated_image = cv2.warpAffine(image1,M,(w1,h1))
cv2_imshow(rotated_image)
cv2.waitKey(0)
```

```
#using imutils to rotate image
image2 = imutils.rotate(image1,-180)
cv2_imshow(image2)
cv2.waitKey(0)
```

```
image3 = imutils.rotate_bound(image1,45)
cv2_imshow(image3)
cv2.waitKey(0)
```

```
blurred_image = cv2.GaussianBlur(image1,(31,31),0)
cv2_imshow(blurred_image)
cv2.waitKey(0)
```

```
image4 = image1.copy()
cv2.rectangle(image4,(100,80),(140,140),(0,255,0),2)
cv2_imshow(image4)
cv2.waitKey(0)
```

```
image5 = image1.copy()
cv2.circle(image5,(120,110),28,(21,234,67),2)
cv2_imshow(image5)
cv2.waitKey(0)
```

```
image6  = image1.copy()
cv2.line(image6,(21,177),(34,211),(133,133,133),5)
cv2.line(image6,(71,107),(134,211),(33,13,3),5)
cv2_imshow(image6)
cv2.waitKey(0)
```
